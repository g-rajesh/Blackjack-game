```python
import random

suits = ['Heart','Queen','Spade','Club']
ranks = ['Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace']
values = {'Two':1,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':10,'Queen':10,'King':10,'Ace':11}

playing = True
```

```python
class Card:
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return self.rank +" of "+self.suit
```

```python
class Deck:
    def __init__(self):
        self.deck = []
        
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit, rank))
                
    def shuffle(self):
        random.shuffle(self.deck)
                
    def __str__(self):
        contents = ''
        for card in self.deck:
            contents += '\n' + card.__str__()
            
        return 'The Deck has' + contents
    
    def single_card(self):
        return self.deck.pop()
```

```python
class Hand():
    
    def __init__(self):
        self.list = []
        self.value = 0
        self.aces = 0
        
    def add_card(self, card):
        self.list.append(card)
        self.value += values[card.rank]
        
        if card.rank == 'Ace':
            self.aces += 1
            
    def adjust_for_aces(self):
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
```

```python
class Chips:
    
    def __init__(self, total=0, bet = 0):
        self.total = total
        self.bet = bet
        
    def win(self):
        self.total += self.bet
        
    def lose(self):
        self.total -= self.bet
```

```python
def take_total(chips):
    
    while True:
        
        try:
            chips.total = int(input('Howmuch do you have in your pocket? '))
        except:
            print('Please enter a valid amount')
            continue
        break
    
```

```python
def take_bet(chips):
    
    while True:
        try:
            chips.bet = int(input('Howmuch would you like to bet for the present game? '))
        except:
            print("It seems like you have not entered an integer value")
        else:
            if chips.bet > chips.total:
                print(f"You have not enough money!!! You have only {chips.total}")
                continue
            break
```

```python
def hit(deck, hand):
    deck.shuffle()
    pulled_card = deck.single_card()
    hand.add_card(pulled_card)
    hand.adjust_for_aces()
```

```python
def hit_or_stand(deck, hand):
    
    global playing
    
    while True:
        
        try:
            choose  = input('Do you want to hit or stand?(h,s): ').lower()
        except:
            print('It seems like you didn\'t enter the character')
            continue
        else:
            if choose == 'h':
                hit(deck, hand)
            elif choose == 's':
                playing = False
            else:
                print('It seems like you haven\'t enter the valid character.Use this(h,s)')
                continue
        break
```

```python
def show_some(player, dealer):
    print('\n')
    print('Dealer Cards:')
    print('One Card Hidden')
    print(dealer.list[0])
    print('\n')
    
    print('Player Cards:')
    for card in player.list:
        print(card)
```

```python
def show_all(player, dealer):
    print('\n')
    print('Dealer Cards:')
    for card in dealer.list:
        print(card)
    print('\n')
    
    print('Player Cards:')
    for card in player.list:
        print(card)
```

```python
def dealer_wins(player,dealer,chips):
    print('\n')
    print('Dealer Wins!!!')
    chips.lose()
    
def dealer_busts(player,dealer,chips):
    print('\n')
    print('Player Wins!!!')
    chips.win()

def player_wins(player,dealer,chips):
    print('\n')
    print('Player Wins!!!')
    chips.win()

def player_busts(player,dealer,chips):
    print('\n')
    print('Dealer Wins!!!')
    chips.lose()
    
def tie(player,dealer):
    print('\n')
    print('Match has been tied!!!PUSH')
```

```python
def replay():
    while True:
        
        try:
            pl_again = input('Do you want to play again? Yes or No? ').lower()
        except:
            print('It seems like you have not entered a string.Use (y,n)')
        else:
            if pl_again[0] == 'y':
                return True
            elif pl_again[0] == 'n':
                return False
            else:
                print('You can be allowed to enter only string. Use (y,n)')
                continue
```

```python
#game starts here

print('Welcome To My BLACKJACK GAME!!!')
print('\n')

player_chips = Chips()
take_total(player_chips)
        

while True:
        deck = Deck()
        deck.shuffle()

        player_hand = Hand()
        dealer_hand = Hand()

        player_hand.add_card(deck.single_card())
        player_hand.add_card(deck.single_card())

        dealer_hand.add_card(deck.single_card())
        dealer_hand.add_card(deck.single_card())
        
        
        take_bet(player_chips)
        
        show_some(player_hand,dealer_hand)
        
        while playing:
            
            hit_or_stand(deck, player_hand)
            show_some(player_hand, dealer_hand)
            
            if player_hand.value > 21:
                player_busts(player_hand,dealer_hand,player_chips)
                break
                
        if player_hand.value <= 21:
            
            while dealer_hand.value < player_hand.value:
                hit(deck,dealer_hand)
                
            if dealer_hand.value > 21:
                dealer_busts(player_hand, dealer_hand, player_chips)
                
            elif dealer_hand.value > player_hand.value:
                dealer_wins(player_hand, dealer_hand, player_chips)
                
            elif dealer_hand.value < player_hand.value:
                player_wins(player_hand, dealer_hand, player_chips)
                
            else:
                tie(player_hand,dealer_hand)
                
        print(f'You have {player_chips.total} now!!!')
        
        if not replay():
            break
```
